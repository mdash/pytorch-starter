```python
data = pd.read_csv('riceClassification.csv')
```

```python
df_x = data.drop(columns = ['id','Class'])
df_y = data['Class']
```

```python
scal = StandardScaler()
x_norm = scal.fit_transform(df_x)
df_x_norm = pd.DataFrame(x_norm, columns=df_x.columns.values)
```

## Autoencoders with continuous data

```python
latent_dims = 2
input_dims = 10
hidden_dims = 4
batch_size = 8192
dropout_p = 0.05

# Tabular data loading and batches here
data = FastTensorDataLoader(torch.from_numpy(df_x_norm.values.astype(np.float32)),
                            torch.from_numpy(df_y.values.astype(np.float32)),
                            batch_size=batch_size, shuffle=True)

autoencoder = Autoencoder(input_dims, hidden_dims, latent_dims, dropout_p).to(device)

autoencoder = train(autoencoder, data, epochs=300, learning_rate=0.3,batch_size=batch_size)
```

```python
emb_x = pd.DataFrame(autoencoder.encoder(torch.from_numpy(df_x_norm.values.astype(np.float32))).to('cpu').detach().numpy(), columns=['dim1','dim2'])
```

```python
df_out = pd.concat([emb_x,df_y], axis=1)
```

```python
df_out.plot(kind='scatter',x='dim1',y='dim2',c=df_out['Class'].values, alpha=0.9, cmap='Dark2');
```
